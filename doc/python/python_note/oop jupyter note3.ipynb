```python
# 属性案例
# 创建Student类，描述学生类
# 学生具有Student.name属性
# 但是name格式并不统一
class Student():
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def intro(self):
        print('hi,my name is {0}'.format(self.name))
s1 = Student('ABB',19)
s2 = Student('bbc',24)
s1.intro()
s2.intro()
```

```python
#规整属性
class Student():
    def __init__(self,name,age):
        self.name = name
        self.age = age
        self.setName(name)
    def setName(self,name):
        self.name = name.upper()
    def intro(self):
        print('hi,my name is {0}'.format(self.name))
    
s1 = Student('Abb',19)
s2 = Student('bbc',24)
s1.intro()
s2.intro()
```

```python
# property案例
# 定义一个Person类，有name，age属性
# 对于任意输入的姓名，我们都希望用大写的方式保存
# 年龄，统一用整数保存
# property(fget,fset,fdel,doc)
class Person():
    '''
    文档
    '''
    def fget(self):
        return self._name*2
    def fset(self,name):
        self._name = name.upper()
    def fdel(self):
        self._name = 'NoName'
    name = property(fget,fset,fdel,'doc') 
```

```python
# __setattr__ 案例
class Person():
    def __init__(self):
        pass
    def __setattr__(self, name, value): #对name这个属性设置
        print(name)
        print('设置属性：{0}'.format(name))
        #死循环，魔法函数在一定条件下自动触发，当对某个属性设置值的时候会触发这个函数，下一行就是对某一属性设置，所以会一直触发这个魔法函数
        print(name)
        self.name = value

p = Person()
print(p.__dict__)
# 对属性进行设置，赋值18
p.age = 18
    
```

```python
# __setattr__ 案例
class Person():
    def __init__(self):
        pass
    def __setattr__(self, name, value): #对name这个属性设置
        
        print('设置属性：{0}'.format(name))
        #死循环，魔法函数在一定条件下自动触发，当对某个属性设置值的时候会触发这个函数，下一行就是对某一属性设置，所以会一直触发这个魔法函数
        # 此种情况，为了避免死循环，规定统一调用父类魔法函数
        ##self.name = value
        super().__setattr__(name,value)

p = Person()
print(p.__dict__)
# 对属性进行设置，赋值18
p.age = 18
    
```

```python
# __gt__
class Student():
    def __init__(self,name):
        self._name = name
    def __gt__(self,obj):
        print('{0}会比{1}大吗'.format(self,obj))
        return self._name > obj._name
stu1 = Student('one')
stu2 = Student('two')

print(stu1 > stu2)
```

```python
# 三种方法的案例
# 实例方法 只有实例可以调用

class Person():
    def eat(self):
        print(self)
        print('eating...')
    # 类方法
    # 类方法的第一个参数，一般命名为cls，区别于self
    @classmethod
    def play(cls):
        print(cls)
        print('playing...')
        
    # 静态方法
    # 不需要用第一个参数表示自身或者类
    @staticmethod
    def say():
        print('saying...')
abb = Person()

#实例用法
abb.eat()
#类方法
Person.play()
abb.play()
#静态方法
Person.say()
abb.say()
```
